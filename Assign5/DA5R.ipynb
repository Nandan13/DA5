```r
df <- read.csv(file="profiles.csv", header=TRUE, sep=",")
```

```r
summary(df)
```

```r
# Subset the attitude data
dat = df[,c(1,19)]

# Plot subset data
plot(dat, main = "% of favourable responses to
     Age and Income", pch =20, cex =2)
```

```r
# Perform K-Means with 2 clusters
set.seed(7)
km1 = kmeans(dat, 5, nstart=100)

# Plot results
plot(dat, col =(km1$cluster +1) , main="K-Means result with 2 clusters", pch=20, cex=2)
```
